```python
# Step 5: Check class distribution
class_counts = df['Class'].value_counts()
plt.figure(figsize=(5,4))
sns.barplot(x=class_counts.index, y=class_counts.values, palette='Set2')
plt.title("Distribution of Fraudulent (1) vs Non-Fraudulent (0) Transactions")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

fraud_ratio = (class_counts[1] / class_counts[0]) * 100
print(f"Fraudulent transactions make up only {fraud_ratio:.3f}% of all transactions.")
```

### Class Distribution Analysis

In this step, we examine the imbalance between legitimate and fraudulent transactions within the dataset.
The visualization below illustrates the proportion of each class, where:

Class 0 → Non-fraudulent transactions

Class 1 → Fraudulent transactions

The chart clearly indicates a significant skew toward non-fraudulent cases, a common characteristic in financial fraud datasets.
Such class imbalance must be carefully addressed during model training to prevent bias toward the majority class.


```python
# Step 6: Distribution of Time and Amount
fig, ax = plt.subplots(1, 2, figsize=(12,4))

sns.histplot(df['Time'], bins=50, ax=ax[0], color='skyblue')
ax[0].set_title("Distribution of Transaction Time")

sns.histplot(df['Amount'], bins=50, ax=ax[1], color='salmon')
ax[1].set_title("Distribution of Transaction Amount")

plt.tight_layout()
plt.show()
```

This step visualizes the distributions of Time and Amount to understand transaction patterns.
Most transactions occur within specific time intervals, and the amounts are highly skewed, with smaller values being more frequent — a typical trend in financial datasets.

```python
# Step 7: Statistical overview of PCA features
pca_features = [col for col in df.columns if col.startswith('V')]
df[pca_features].describe().T.head(10)
```

The dataset contains PCA-transformed features (V1–V28), which are linear combinations of original variables ensuring confidentiality.
This step provides a statistical summary (mean, standard deviation, min, max) to understand the distribution and scale of these components.

```python
# Step 8: Correlation heatmap
plt.figure(figsize=(10,8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

This step visualizes the correlation matrix of all numerical features to identify interdependencies and multicollinearity within the dataset.
Highly correlated features may carry redundant information, which can affect model performance and interpretation.

```python
# Step 9: Detect potential outliers in 'Amount'
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Amount'])
plt.title("Boxplot of Transaction Amounts")
plt.show()
```

A boxplot of the Amount feature is used to detect outliers and understand the spread of transaction values.
Most transactions are of low value, while a few high-value outliers may indicate potential fraudulent activity.
This visualization highlights skewness and extreme values, helping in scaling and outlier treatment during preprocessing.
