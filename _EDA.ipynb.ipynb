```python
# Notebook 2: Exploratory Data Analysis (EDA)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8, 5)

# Test plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Test Plot")
plt.show()

```

This notebook performs a comprehensive Exploratory Data Analysis to understand the structure, distribution, and relationships within the dataset.
EDA helps identify trends, anomalies, and patterns that guide data preprocessing and model development.
This initialization sets up all essential libraries and visualization styles to ensure consistency across the analysis.

```python
# Summary statistics
df.describe().T
```

This step provides a statistical overview of all numerical features, including measures such as mean, standard deviation, and quartiles.
It helps identify data distribution, scale variations, and potential outliers across features.
These descriptive statistics form the foundation for deeper exploratory insights in subsequent steps.

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df['Time'], bins=50, ax=ax[0], color='skyblue')
ax[0].set_title("Distribution of Transaction Time")

sns.histplot(df['Amount'], bins=50, ax=ax[1], color='salmon')
ax[1].set_title("Distribution of Transaction Amount")

plt.tight_layout()
plt.show()
```

This visualization explores how transaction time and transaction amount are distributed across the dataset.
Most transactions occur within specific time ranges, while transaction amounts are right-skewed, indicating many low-value transactions and a few high-value outliers.

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Amount'], color='lightcoral')
plt.title("Boxplot of Transaction Amounts")
plt.show()
```

A boxplot of the Amount feature is used to visualize the spread and detect potential outliers in transaction values.
The plot shows that most transactions involve low amounts, while a few high-value outliers may represent unusual or fraudulent activities.

```python
# Select PCA-based features
pca_features = [col for col in df.columns if col.startswith('V')]
df[pca_features].describe().T.head(10)
```

The dataset includes anonymized PCA-transformed features (V1–V28), representing key transaction characteristics after dimensionality reduction.
Descriptive statistics below summarize their distributions and variability.
This analysis helps identify feature scales, variances, and potential anomalies across principal components.

```python
plt.figure(figsize=(10, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

A correlation heatmap visualizes relationships between all features, highlighting multicollinearity and patterns that may influence model performance.
Strong positive or negative correlations can indicate redundant features or hidden dependencies.

```python
class_counts = df['Class'].value_counts()
plt.figure(figsize=(5,4))
sns.barplot(x=class_counts.index, y=class_counts.values, palette='Set2')
plt.title("Fraudulent (1) vs Non-Fraudulent (0) Transactions")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

fraud_ratio = (class_counts[1] / class_counts[0]) * 100
print(f"Fraudulent transactions make up only {fraud_ratio:.3f}% of all transactions.")
```

The dataset is highly imbalanced, with very few fraudulent transactions compared to non-fraudulent ones.
⚠️ Insight: Fraudulent cases form a very small percentage of total transactions, indicating a strong class imbalance that needs to be handled before modeling.

```python
# Compare amount distribution for fraud vs non-fraud
plt.figure(figsize=(6,4))
sns.boxplot(x='Class', y='Amount', data=df, palette='Set3')
plt.title("Transaction Amounts by Class")
plt.show()
```

This visualization compares transaction amounts between fraudulent and non-fraudulent transactions.
Insight: Fraudulent transactions often show distinct amount patterns compared to normal ones, which can help models learn differentiating features.

```python
# Correlation of each feature with target variable
corr_with_target = df.corr()['Class'].sort_values(ascending=False)
corr_with_target.head(10)
```

This step identifies the features most correlated with the target variable (Class), highlighting which components contribute most to fraud detection.
Insight: Features showing higher positive or negative correlations with Class are strong predictors and can be prioritized during model training.
